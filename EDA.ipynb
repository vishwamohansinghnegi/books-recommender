```python
df = books[
    ~(books['authors'].isnull()) &
    ~(books['categories'].isnull()) &
    ~(books['thumbnail'].isnull()) &
    ~(books['description'].isnull()) &
    ~(books['average_rating'].isnull())    # num_pages and ratings_count also missing same rows as average_rating
]

df.isnull().sum()
```

## Analyzing each column

```python
df.value_counts('categories').reset_index()

plt.figure(figsize=(15,20))
sns.countplot(df['categories'], color='orange')

# we can see that there are only few major categories and other categories have only one or two books only
```

## Description of smaller length are of no use!

```python
df['len_of_description'] = df['description'].str.split().str.len()
df['len_of_description']
```

```python
max(df['len_of_description'])
```

```python
sns.histplot(df['len_of_description'])
```

```python
df.loc[df['len_of_description']<25, 'description']

# Tried on multiple values and decided 25 as optimal value for thershold,means we need atleast 25 words for some considerable description
```

```python
df = df[df['len_of_description']>=25]
df.shape
```

```python
df.head()
```

```python
# We can see that subtitles are not providing that much good info so we are going to club it with title
import numpy as np

df['title_subtitle'] = np.where(
    df['subtitle'].isnull(), df['title'],
    df['title'] + ': ' + df['subtitle']
)

df.head()
```

```python
# We need unique identifier for each description, so we are going to add isbn10 with description for this
# as when we will get result from model it will be decription so we will fetch isbn10 and find the book title out of there

df['tagged_description'] = df['isbn13'].astype(str) + ' ' + df['description']
df.head()
```

```python
# Dropping irrelevant cols and saving the cleaned data

df.drop(['subtitle', 'len_of_description'], axis=1, inplace=True)
df.to_csv('books_cleaned.csv', index=False)
```
