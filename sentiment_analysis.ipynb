# Classifying books into different sentiments based on their description

```python
import pandas as pd

df = pd.read_csv('books_categories.csv')
df.head()
```

```python
from transformers import pipeline

classifier = pipeline('text-classification', model='j-hartmann/emotion-english-distilroberta-base', top_k=None)
```

```python
# Processing whole description has 2 limitations. 1st is token limit might be there and 2nd for each sent pred will be better
# So now we will process sentence-wise, sentiment of each sentence and then average it for sentiment

import numpy as np

# All possible emotion labels that the model can output.
emotion_labels = ["anger", "disgust", "fear", "joy", "sadness", "surprise", "neutral"]

# Dictionaries and lists to store the final results.
isbn = []
emotion_scores = {label: [] for label in emotion_labels}


# --- 2. NEW FUNCTION: Calculate Average Scores ---

def calculate_avg_emotion_scores(predictions):
    """
    Processes model predictions for a list of sentences and calculates the
    average score for each emotion.
    """
    # Initialize a dictionary to hold a list of scores for each emotion.
    per_emotion_scores = {label: [] for label in emotion_labels}

    # The 'predictions' input is a list of lists (one list per sentence).
    # e.g., [[{'label':'joy', 'score':0.9}, ...], [{'label':'sadness', 'score':0.8}, ...]]

    # Loop through the predictions for each sentence
    for sentence_prediction_list in predictions:
        # Loop through each individual emotion dictionary in the list
        for emotion in sentence_prediction_list:
            label = emotion['label']
            score = emotion['score']
            # Append the score to the correct list using the label name
            if label in per_emotion_scores:
                per_emotion_scores[label].append(score)

    # Calculate the average score for each emotion.
    # If a description had no sentences, default the score to 0.0.
    avg_scores = {
        label: float(np.mean(scores)) if scores else 0.0
        for label, scores in per_emotion_scores.items()
    }

    return avg_scores


# --- 3. MAIN LOOP (Modified to call the new function) ---

# Processing the first 10 books as an example.
for i in range(3):
    # Store the ISBN
    isbn.append(df["isbn13"][i])

    # Split the description into sentences (a simple split on '.' is used here)
    sentences = df["description"][i].split(".")
    # Filter out any empty strings that may result from the split
    sentences = [s for s in sentences if s.strip()]

    predictions = classifier(sentences, top_k=None) # top_k=None ensures all scores are returned
    # Calculate the average scores using the new function
    avg_scores = calculate_avg_emotion_scores(predictions)


    # Append the calculated average scores for this book to the main dictionary
    for label in emotion_labels:
        emotion_scores[label].append(avg_scores[label])

emotion_scores
# At the end of this loop, the 'emotion_scores' dictionary will contain lists of
# the average emotion scores for each of the 10 book descriptions.
```

```python
# Now Processing all books
from tqdm import tqdm

isbn = []
emotion_scores = {label: [] for label in emotion_labels}

for i in tqdm(range(len(df))):
    # Store the ISBN
    isbn.append(df["isbn13"][i])

    # Split the description into sentences (a simple split on '.' is used here)
    sentences = df["description"][i].split(".")
    # Filter out any empty strings that may result from the split
    sentences = [s for s in sentences if s.strip()]

    predictions = classifier(sentences, top_k=None) # top_k=None ensures all scores are returned
    # Calculate the average scores using the new function
    avg_scores = calculate_avg_emotion_scores(predictions)


    # Append the calculated average scores for this book to the main dictionary
    for label in emotion_labels:
        emotion_scores[label].append(avg_scores[label])

emotion_scores
# At the end of this loop, the 'emotion_scores' dictionary will contain lists of
# the average emotion scores for each of the 10 book descriptions.
```

```python
emotions_df = pd.DataFrame(emotion_scores)
emotions_df['isbn13'] = isbn
emotions_df
```

```python
df = pd.merge(df, emotions_df, on='isbn13')
df
```

```python
df.to_csv('books_sentiment.csv', index=False)
```
