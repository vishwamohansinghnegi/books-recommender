```python
import pandas as pd

df = pd.read_csv('books_cleaned.csv')
df.head()
```

```python
df['categories'].value_counts().reset_index()[:12]
# Considering top 12 categories
```

```python
# We can broadly classify categories as fiction and non-fiction according to the above categories

category_mapping = {'Fiction' : "Fiction",
 'Juvenile Fiction': "Children's Fiction",
 'Biography & Autobiography': "Nonfiction",
 'History': "Nonfiction",
 'Literary Criticism': "Nonfiction",
 'Philosophy': "Nonfiction",
 'Religion': "Nonfiction",
 'Comics & Graphic Novels': "Fiction",
 'Drama': "Fiction",
 'Juvenile Nonfiction': "Children's Nonfiction",
 'Science': "Nonfiction",
 'Poetry': "Fiction"}

df["simple_categories"] = df["categories"].map(category_mapping)
df.head()
```

```python
df[~(df['simple_categories'].isnull())].shape, df.shape

# now most of the  books have simple categories and 1300 books dont have
```

```python
from transformers import pipeline

fiction_categories = ['Fiction', 'Nonfiction']

pipe = pipeline(
    'zero-shot-classification',
    model='facebook/bart-large-mnli'
)
```

```python
seq = df.loc[df['simple_categories'] == 'Fiction', 'description'].reset_index(drop=True)[0]
```

```python
pipe(seq, fiction_categories)

# We can see that our book was fiction and our model is also predicting the same
```

```python
# Fetching label

import numpy as np

res = pipe(seq, fiction_categories)
max_index = np.argmax(res['scores'])
pred_label = res['labels'][max_index]
pred_label
```

```python
# Now making a function for the same

def generate_pred(seq, fiction_categories):
    res = pipe(seq, fiction_categories)
    max_index = np.argmax(res['scores'])
    pred_label = res['labels'][max_index]
    return pred_label
```

```python
actual_cat = list()
pred_cat = list()

sequence = df.loc[df['simple_categories'] == 'Fiction', 'description'].reset_index(drop=True)
for i in range(100):
    seq = sequence[i]
    pred_cat += [generate_pred(seq, fiction_categories)]
    actual_cat += ['Fiction']
```

```python
sequence = df.loc[df['simple_categories'] == 'Nonfiction', 'description'].reset_index(drop=True)
for i in range(100):
    pred_cat += [generate_pred(sequence[i], fiction_categories)]
    actual_cat += ['Nonfiction']
```

```python
pred_df = pd.DataFrame({'actual_categories':actual_cat, 'predicted_categories':pred_cat})
pred_df
```

```python
pred_df['correct_pred'] = np.where(pred_df['actual_categories']==pred_df['predicted_categories'], 1, 0)
pred_df
```

```python
print(f'Accuracy: {pred_df['correct_pred'].sum()/len(pred_df)}')
# accuracy is good so we can appply it to all the remaining rows
```

```python
# Now for predicting categories of missing labels, so that we can merge them back into dataset
from tqdm import tqdm

isbns = []
pred_cats = []

missing_cats = df.loc[df['simple_categories'].isnull(), ['isbn13', 'description']].reset_index(drop=True)

for i in tqdm(range(len(missing_cats))):
    seq = missing_cats['description'][i]
    pred_cats += [generate_pred(seq, fiction_categories)]
    isbns += [missing_cats['isbn13'][i]]
```

```python
missing_pred_df = pd.DataFrame({'isbn13':isbns, 'predicted_categories': pred_cats})
missing_pred_df
```

```python
df = pd.merge(df, missing_pred_df, on='isbn13', how='left')
df['simple_categories'] = np.where(df['simple_categories'].isnull(), df['predicted_categories'], df['simple_categories'])
df.drop(columns=['predicted_categories'], inplace=True)
```
